```python
import networkx as nx
from ipysigma import Sigma, SigmaGrid
import pelote
```

```python
label_palette = {
    "maison edition": "#cda64b",
    "librairie": "#678f39",
    "association": "#8568d5"
}
```

```python
node_palette = {
    "maison edition": "#d4d17f",
    "librairie": "#a7d6ca",
    "association": "#aac7e3"
}
```

```python
# Importing a gexf graph
g = nx.read_gexf('/home/bmazoyer/Downloads/Genre - Associations et maisons edition.gexf')
pelote.crop_to_largest_connected_component(g)

# Displaying the graph with a size mapped on degree and
# a color mapped on a categorical attribute of the nodes
widget = Sigma(g,
    node_label_size=g.in_degree, 
    node_size=g.degree, 
    label_rendered_size_threshold=1, 
    label_density=1, 
    label_grid_cell_size=75, 
    node_color="tag_Categorie", 
    node_color_palette=node_palette,  
    node_label_color="tag_Categorie", 
    node_label_color_palette=label_palette
)
```

```python
widget
```

```python
Sigma.write_html(
    g, 
    "asso_edition_librairie.html", 
    fullscreen=True,
    node_label_size=g.in_degree, 
    node_size=g.degree, 
    label_rendered_size_threshold=1, 
    label_density=1, 
    label_grid_cell_size=75, 
    node_color="tag_Categorie", 
    node_color_palette=node_palette,  
    node_label_color="tag_Categorie", 
    node_label_color_palette=label_palette
)
```
